```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np
```

```python
data={
    "Hours":[1,2,3,4,5,6],
    "Score":[52,57,65,70,75,80]
}
df=pd.DataFrame(data)
df.to_csv("Student.csv")
```

```python
df=pd.read_csv("Student.csv")
x=df[["Hours"]]
y=df["Score"]

model=LinearRegression()
model.fit(x,y)

predition_score=model.predict(x)

mae=mean_absolute_error(y,predition_score)
mse=mean_squared_error(y,predition_score)
rmse=np.sqrt(mse)

# Show result 
print("Man absolute error  : ",mae)
print("Mean Squared error : ",mse)
print("Root mean squareed error    :",rmse)

hour=float(input("Enter a hour"))
new_pred=model.predict([[hour]])[0]
print("New Prediction : ",new_pred)

```
