## Regression : 
### Linear Regression :
Prediction of number using straight line
1. Find Pattern
2. Strainght Line
3. prediction using the line
> y=mx+b , y= output x=input





```python
from sklearn.linear_model import LinearRegression
X=[[1],[2],[3],[4],[5]]
Y=[40,50,65,75,90]
model=LinearRegression()
model.fit(X,Y)
hours=float(input("Enter the hours you studied : ",))
# [[value]] is the always in 2d , because 
predicted_marks=model.predict([[hours]])

print(f"Based on your hourse {hours} , you may score {predicted_marks}")
```

### Classification model : 


1. Logistic Regression : 
                       In classification models , we don't predict the random number . but choose something.eg: pass,fail etc


```python
from sklearn.linear_model import LogisticRegression
x=[[1],[2],[3],[4],[5]]
y=[0,0,0,1,1]# 0= fail , 1=pass

model=LogisticRegression()
model.fit(x,y)
hours=float(input("Enter the hours you have studied : "))
predicted_result=model.predict([[hours]])[0]

if predicted_result==0:
    print(f"Based on your hours{hours} , You will fail")
elif predicted_result==1:
    print(f"Based on your hours {hours} , You are going to pass")
```

2. KNN : It makes decisions based on neighbours , for example , if 4 out of 5 people are going to gym , then the rest will also thinking of the same . 
also called lazy model.So practically i makes decision using the nearest dataset 

* choose K value , means no of neibours to be considered .
* Find the closest neighbour
* How many are in each class 
> Always sue odd nukbers for KNN 
* > Alwys avoid for a big data

```python
# model to classify fruit based on weight,size
from sklearn.neighbors import KNeighborsClassifier

x=[
    [180,7],
    [200,7.5],
    [250,8],
    [300,8.5],
    [330,9],
    [360,9.5]
]

#0=Apple , 1=orange

y=[0,0,0,1,1,1]

model=KNeighborsClassifier(n_neighbors=3)

model.fit(x,y)

weight=float(input("Enter the  weight of fruit in grams : "))
size=float(input("Enter the size in CM"))

prediction=model.predict([[weight,size]])[0]# writing 0 in single bracket , because predict() returns the list of one elements
print(prediction)

if prediction==0:
    print("Likely an apple")
else:
    print("likely an orange")
```

3. Desicion Tree : 
For example , doctor asking the patient that weather he have ccough and fever to predict the flue's presence . 
> Defination : As a defination i can say that this model classifies based on some other data and then learn from that data to make the prediction .

* it will work better if follow clear steps 
* trees can be too smart , will result in overfitting
* easy to explain to anyone 


```python
from sklearn.tree import DecisionTreeClassifier
x=[#size,colorshade
    [7,2],
    [8,3],
    [9,8],
    [10,9]
]
y=[0,0,1,1]

model=DecisionTreeClassifier()

model.fit(x,y)

size=float(input("Enter the size of the fruit : "))

color=float(input("Enter the color shade of fruit(1-10):"))

result=model.predict([[size,color]])[0]

if result==0:
    print("this is an apple")
else:
    print("Likely an orange")
```

.fit() is a training step for the data , we need to pass x,y here 
.predict([[value]]) its work is simply to predict based on the value given by user .

fit is use only once , but predict can be used many times .
