## Classification Matrix : 
                            Learn to evaluation classification models using the accurace , precision , recall , and F1-Score .
*  Accuracy : correct_prediction/total_Prediction
* precision : used to avoid the false predictions , things that were preddicted but was wrong . 
* Recall : thigs that were True but  not predicted . 
* F1-Score: to macth the recall , and presicion .



```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

#True answers (what actually happened)
y_true=[1,0,1,1,0,1,0]

# predictions
y_pred=[1,0,1,0,0,1,1]

print("Accurace ",accuracy_score(y_true,y_pred))
print("precision score :  ",precision_score(y_true,y_pred))
print("recall_score ",recall_score(y_true,y_pred))
print("f1_score",f1_score(y_true,y_pred))
```
