## Encoding Categorical Data
### We will be doing two major things :
1. Label encoding
2. one-hot enconding 

### Label encoding
1. every catttegorry to be changedd in a unique number
 for example : male=0,female =1
2. this tyoe of encoding to b eussed , where there mostly binarry type of values .  for example : Gender , married or un-married

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

df=pd.read_csv("machine.csv")
df_label=df.copy()

le=LabelEncoder()
df_label["STATUS"]=le.fit_transform(df_label["STATUS"])
print("Label encoded data")
print(df_label[["STATUS"]])
df_label.to_csv("Labledmachien.csv")
```
