### Missing Data Handelling : 
* We  will be using pandas library for it .
* Why : ML model can not handel the missing or NaN values .
> Top three method for it : 
1. dropna() . For removing any row or column .
2. fillna(value) , for filling any null value .
3. isnull().sum() , for calculating how much NaN values are there in data .

```python
import pandas as pd
```

```python
data={
    "Name":["pavan","Kapil","Lalit","Ishan","Om"],
    "Age":[25,None,44,23,None],
    "Salary":[5000,6000,7000,None,None]

}

df=pd.DataFrame(data)
print("Original data Fram e")
print(df)

#This is the way to drop the row completely , where the info is incomplete
print("Total missing values",df.isnull().sum())
df_drop=df.dropna()
print(df_drop)
```

```python
# way to with aaverage value , this fills the values by columns 
df["Age"].fillna(df["Age"].mean(),inplace=True)
df["Salary"].fillna(df["Salary"].mean(),inplace=True)
print(df)
```

> Advices : 
1. Never try  to drop the data at first , always try to fill it 
2. When there is numeric data try to fill it with the mean and median . And when there is categorial data try to fill it with mode .
3. Always try to see how much percent of data is misssing before cleaning .


```python
# Way to see ho much data is missing 
print(df.isnull().mean()*100)
# it will show how much is missing in every data
```
