```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import pandas as pd
```

```python
df=pd.read_csv("Dataset.csv")
df2=df.copy()

columns_to_clean=["Peak","All Time Peak","Actual gross","Adjusted gross (in 2022 dollars)","Average gross"]
for col in columns_to_clean:
    #to clean everything which is bitween the columns
    df2[col] = df2[col].replace(r'\[.*?\]', '', regex=True)
    #to clean everything except the numeric digits
    df2[col]=df2[col].replace(r'\D',"",regex=True)
    # convert everything to numeric values again
    df2[col]=pd.to_numeric(df2[col])

df2["Ref."]=df2["Ref."].replace(r'\D','',regex=True)
print(df2)
```

```python
df2.isnull().sum()
```

```python
peak_mode=df2["Peak"].mode()[0]

df2["Peak"].fillna(peak_mode,inplace=True)
df2["All Time Peak"].fillna(df2["All Time Peak"].mean(),inplace=True)

print(df2)
df2.isnull().sum()
```
