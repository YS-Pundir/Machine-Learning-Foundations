```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import pandas as pd
```

```python
df=pd.read_csv("Dataset.csv")
df2=df.copy()
```

```python
print(df2)
```

```python
column_to_clean=["Peak","All Time Peak","Actual gross","Adjusted gross (in 2022 dollars)","Average gross"]

for col in column_to_clean:
    df2[col]=df2[col].str.replace(r'\[.*?\]',"",regex=True)

    df2[col]=df2[col].str.replace(r'\D',"",regex=True)

    df2[col]=pd.to_numeric(df2[col])

    
df2=df2.drop("Ref.",axis=1)
```

```python
print(df2)
```

```python
df2.to_csv("Cleaned_Dataset.csv")
```

```python
print("Total NaN data in the Dataset")
df2.isnull().sum()
# first we will try to fill it 
df2["Peak"].fillna(df2["Peak"].mode()[0],inplace=True)
df2["All Time Peak"].fillna(df2["All Time Peak"].mean(),inplace=True)
df2.to_csv("Cleaned_Dataset.csv")
```

```python
df2.drop("Tour title",axis=1)
```

```python
df2=pd.get_dummies(df2,columns=["Artist"],dtype=int)
df2.to_csv("Cleaned_Dataset.csv")

df2=df2.drop("Tour title",axis=1)
df2=df2.drop("Year(s)",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scaled=scalar.fit_transform(df2)

df3=pd.DataFrame(scaled,columns=df2.columns)
print(df3)

df3.to_csv("Cleaned_Dataset.csv")

```

```python
from sklearn.model_selection import train_test_split

x=df3.drop("Actual gross",axis=1)
y=df3["Actual gross"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("Training data")
print(x_train)
print("Teesting data")
print(x_test)
print("training outcome")
print(y_test)
```
