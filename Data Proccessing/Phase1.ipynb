```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import pandas as pd
```

```python
df=pd.read_csv("Dataset.csv")
df2=df.copy()

columns_to_clean=["Peak","All Time Peak","Actual gross","Adjusted gross (in 2022 dollars)","Average gross","Ref."]
for col in columns_to_clean:
    #to clean everything which is bitween the columns
    df2[col] = df2[col].replace(r'\[.*?\]', '', regex=True)
    #to clean everything except the numeric digits
    df2[col]=df2[col].replace(r'\D',"",regex=True)

print(df2)
```
