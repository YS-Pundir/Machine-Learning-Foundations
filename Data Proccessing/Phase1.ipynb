```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import pandas as pd
```

```python
df=pd.read_csv("Dataset.csv")
df2=df.copy()
```

```python
print(df2)
```

```python
column_to_clean=["Peak","All Time Peak","Actual gross","Adjusted gross (in 2022 dollars)","Average gross","Ref."]

for col in column_to_clean:
    df2[col]=df2[col].str.replace(r'\[.*?\]',"",regex=True)

    df2[col]=df[col].str.replace(r'\D',"",regex=True)

    df2[col]=pd.to_numeric(df2[col])
```

```python
print(df2)
```

```python
df2.to_csv("Cleaned_Dataset.csv")
```
