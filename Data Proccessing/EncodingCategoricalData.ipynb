## Encoding Categorical Data
### We will be doing two major things :
1. Label encoding
2. one-hot enconding 
* If there is just one type of text data then we don't apply encodding as machine learnign learn nothing from it .

### Label encoding
* As model can not work on the text data , it needed to be in numerical data .this is done with the help of encoding. 
1. every cattegory to be changed in a unique number
 for example : male=0,female =1
2. this tyoe of encoding to b eussed , where there mostly binary type of values .  for example : Gender , married or un-married

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
df=pd.read_csv("Cleaned_machine.csv")
df_label=df.copy()  # making a copy fo the data form csv file  

df_label=df_label.drop(columns=["UNITS","Magnitude","Subject","Group","Series_title_1"],axis=1)#  the drop returns a new copy of data 
print("Data After deleting the same values columns ")
print(df_label)
le=LabelEncoder()


df_label["STATUS_Labeled"]=le.fit_transform(df_label["STATUS"])
print("The labeled data is : ")
print(df_label)


```

### one-Hot encoding :
 This encoding is used to ecode when we have more then two categroy of the data 
 

```python
df_label=pd.get_dummies(df_label,columns=["Series_reference","Series_title_2","Series_title_3"],dtype=int)
print("one hot encoded data")
print(df_label)
df_label.to_csv("Cleaned_machine.csv")
```
